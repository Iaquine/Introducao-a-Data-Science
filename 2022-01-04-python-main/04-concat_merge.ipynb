# merging

```python
person = pd.read_csv('data/survey_person.csv') 
site = pd.read_csv('data/survey_site.csv') 
survey = pd.read_csv('data/survey_survey.csv') 
visited = pd.read_csv('data/survey_visited.csv')
```

```python
person
```

```python
survey
```

```python
# one to one merge same as o2o

visited_sub = visited.loc[[0, 2, 6]] 
visited_sub
```

```python
site
```

```python
o2o = visited_sub.merge(site, left_on=['site'], right_on=['name'])
o2o
```

```python
visited_sub.shape
```

```python
o2o.shape
```

```python
# many to one the same as m2o
m2o = site.merge(visited, left_on='name', right_on='site')
m2o
```

```python
df1 = pd.DataFrame({
    'a': ['a1', 'a2', 'a3', 'a4', 'a5'],
    'b': [1, 1, 2, 2, 3]
})

df2 = pd.DataFrame({
    'a2': ['2a1', '2a2', '2a3', '2a4', '2a5'],
    'b2': [1, 1, 1, 2, 2]
})
```

```python
m2m = pd.merge(df1, df2, left_on='b', right_on='b2')
m2m
```
