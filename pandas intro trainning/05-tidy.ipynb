```python
ebola = pd.read_csv('data/country_timeseries.csv')
```

```python
ebola
```

```python
ebola_long = ebola.melt(id_vars=['Date','Day'])
ebola_long
```

```python
variable_split = ebola_long['variable'].str.split('_') #it is a series
```

```python
type(variable_split.iloc[0])
```

```python
variable_split.iloc[0][0]
```

```python
status_variable = variable_split.str.get(0)
status_variable
```

```python
country_variable = variable_split.str.get(1)
country_variable
```

```python
ebola_long['case_death']= status_variable
ebola_long['country'] = country_variable
```

```python
ebola_dropped = ebola_long.drop('variable', axis=1)
```

```python
ebola_dropped
```

```python
ebola_tidy = (ebola_dropped
              .pivot_table(index=['Date', 'Day', 'country'], 
                           columns='case_death', 
                           values='value')
              .reset_index())
```
