```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
type(tips)
```

```python
tips.plot(kind = 'scatter', x = 'total_bill', y = 'tip')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='sex', data=tips, hue ='time')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips, hue ='time')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips, col ='time')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips, col ='time', hue='sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips, col ='time', hue='sex', row='smoker')
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```
